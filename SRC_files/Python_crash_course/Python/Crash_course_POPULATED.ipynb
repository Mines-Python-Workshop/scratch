## 2) Fitting data using least squares

```python
# scipy is scientific python and contains more advanced functionalities
# such as least squares, curve fitting, and linear regression
from scipy.optimize import leastsq
from scipy import integrate
from scipy.optimize import curve_fit
from scipy.stats import linregress
```

### 2a) Simple Example 1: Linear Regression using scipy linregress

```python
# Example: Load external data (crash_course_data1.dat) and fit to a linear regression
x,y = np.loadtxt('./data/crash_course_data1.dat',
                 unpack=True,
                 dtype=float,
                 delimiter=',')
```

```python
# SciPy has a built-in linear regression
slope, intercept, r_value, p_value, stderr = linregress(x,y)

# Print the fitted values and R-squared
print("Slope: %1.5f" % slope)
print("Itercept: %1.5f" % intercept)
print("R value: %1.5f" % r_value)

#Note: Above is a nice way to print data with labels.
# "1.5f" is a formatting option;
# f=float
# 1.5 = 1 digit before the decimal, 5 digits after the decimal
```

```python
# Plot the data and the fit on one graph. There are three options:

# Option 1: Write a linear function and pass m and b
def line(x,m,b): #x is the argument, m is slope, b = intercept
    return m*x + b

y_fit = line(x,slope,intercept)

plt.figure()
plt.plot(x,y,label='data',marker='o',linestyle=None,color="black")
plt.plot(x,y_fit,label='fit',color='red')
```

```python
# Option 2: Define a new list called y_fit and plot it

y_fit = slope*x+intercept

plt.figure()
plt.plot(x,y,label='data',marker='o',linestyle=None,color="black")
plt.plot(x,y_fit,label='fit',color='red')
```

```python
# Option 3: Do the arithmetic in the plt.plot() command

plt.figure()
plt.plot(x,y,label='data',marker='o',linestyle=None,color="black")
plt.plot(x,slope*x+intercept,label='fit',color='red')
```

### 2b) Challenging Example: Fit data to a written function with fitted parameters

```python
# Load "crash_course_data3.dat" using np.loadtxt
x,y = np.loadtxt("./data/crash_course_data3.dat",dtype=float,unpack=True,delimiter=' ')
```

```python
# Plot the data - can we guess the functional form?
plt.figure()
plt.plot(x,y)
```

```python
# Define a function with variables to fit the data
## By looking at the data, we can guess that it's following a parabola

def fit(x,a,b,c):  #a, b, c are fitted parameters
    y = a*x**2 + b*x + c
    return y
```

```python
init = [1,2,3] #initial values for a, b, and c;

## Use scipy curve_fit
### Fit Coef = fitted coefficients
### CoefsCov = covariance of the coefficients
### p0 = initial parameters
### Bounds (tuple) sets the lower and upper bounds for all parameters
FitCoef, CoefsCov = curve_fit(fit, x, y, p0=init,bounds=(0,np.inf)) 

## CoefsErr = error in the fitted parameters
CoefsErr = np.sqrt(np.diag(CoefsCov))

## Write out fitted values by plugging coefficients
## into the arguments for the fit function
y_fit=fit(x,FitCoef[0],FitCoef[1],FitCoef[1])

##Print the values
print ("a: %1.2f +/- %2.2e" % (FitCoef[0],CoefsErr[0]))
print ("b: %1.2f +/- %2.2e" % (FitCoef[1],CoefsErr[1]))
print ("c: %1.2f +/- %2.2e" % (FitCoef[2],CoefsErr[2]))

#Plot data and fit
plt.figure(figsize=(6,6))
plt.plot(x,y,'o',color='black',label='data') #Plot data x vs. y
plt.plot(x,y_fit,color='orange',label='fit') #Plot x vs. y_fit

# Put some text on the plot
plt.text(0.25, 3.75E5, 'fit = ax$^2$ + bx + c', horizontalalignment='left', fontsize=15, color='black')
plt.text(0.25, 3.5E5, 'a =  %1.2f +/- %2.2e' % (FitCoef[0],CoefsErr[0]), horizontalalignment='left', fontsize=12, color='black')
plt.text(0.25, 3.25E5, 'b =  %1.2f +/- %2.2e' % (FitCoef[1],CoefsErr[1]), horizontalalignment='left', fontsize=12, color='black')
plt.text(0.25, 3E5, 'c =  %1.f +/- %2.2e' % (FitCoef[2],CoefsErr[2]), horizontalalignment='left', fontsize=12, color='black')

plt.legend()
```

### <span style='color:Purple'> Practice: Load crash_course_data4.dat. Fit the data to a Gaussian function using curve_fit. Plot another trace that is the difference curve between data and fit.</span> 

### <span style='color:Purple'> Hint: Here's the Gaussian function :) a = scale factor, b = center of peak, c = defines peak width</span>  
<img src="./images/Gaussian_img.png" width="300"/>

```python
# Load data
x,y = np.loadtxt("./data/crash_course_data4.dat", dtype=float, unpack=True)
```

```python
# Quick plot to make sure data loaded properly
# and also to guess good starting values

plt.figure(figsize=(8,8))
plt.plot(x,y)
```

```python
#Define Gaussian function

def gaussian(x,a,b,c):
    epnt = -((x-b)**2)/(2*c**2) # To keep track of parentheses and PEMDAS, 
                                # I defined the term in the exponent separately
    g = a * np.exp(epnt)
    return g
```

```python
# Initial guess for fitted parameters
init = [2.5E4,3,0.8] #Initial guess for a, b, and c

# Define curve-fitting commands
FitCoef, CoefsCov = curve_fit(gaussian, x, y, p0=init,bounds=((0,np.inf))) #We can demand bounds
CoefsErr = np.sqrt(np.diag(CoefsCov))
gauss_fit=gaussian(x,FitCoef[0],FitCoef[1],FitCoef[2])


##Print the values
print ("a: %1.2f +/- %2.2e" % (FitCoef[0],CoefsErr[0]))
print ("b: %1.2f +/- %2.2e" % (FitCoef[1],CoefsErr[1]))
print ("c: %1.2f +/- %2.2e" % (FitCoef[2],CoefsErr[2]))
```

```python
# Plot data, fit, and difference curve
plt.figure(figsize=(8,8))
plt.plot(x,y,'o',color='black')
plt.plot(x,gauss_fit,'-',color='orange')
plt.plot(x,y-gauss_fit-10000,'-', color='grey') #Difference curve, offset by -10000 in y
```
